1. What Is an Array?

An array is a collection of elements of the same data type stored in contiguous memory locations and accessed using an index.


2. Static Array

A static array has a fixed size that is decided at compile time and cannot be changed during program execution.

Key Characteristics

- Size is fixed
- Memory is allocated once
- Stored usually in stack memory
- Fast access
- Risk of wasted memory or overflow-


Advantages

- Simple to use
- Faster (no resizing overhead)
- Less memory fragmentation
- Cache-friendly

Disadvantages

- Size is fixed
- Memory may be wasted
- Overflow if you exceed size
- Not suitable when size is unknown

Use Cases

- Small programs
- Embedded systems
- When size is known beforehand
- Performance-critical code

```python
static_array = [1, 2, 3, 4, 5]
```

3. Dynamic Array

A dynamic array can grow or shrink at runtime. Memory is allocated dynamically (usually on the heap).

Key Characteristics

- Size can change
- Memory allocated at runtime
- Stored in heap memory
- Needs manual management (in low-level languages)

How Dynamic Arrays Grow

When the array is full:
1. Allocate new memory (usually double size)
2. Copy old elements
3. Free old memory
4. Insert new element

Advantages

- Flexible size
- Efficient memory usage
- Suitable when size is unknown
- Used in modern programming languages

Disadvantages

- Slower than static arrays (resizing cost)
- More complex memory handling
- Possible memory fragmentation


4. Static Array vs Dynamic Array (Comparison)

| Feature      | Static Array | Dynamic Array                 |
| ------------ | ------------ | ----------------------------- |
| Size         | Fixed        | Resizable                     |
| Allocation   | Compile time | Runtime                       |
| Memory       | Stack        | Heap                          |
| Speed        | Faster       | Slightly slower               |
| Memory Usage | Can waste    | Efficient                     |
| Flexibility  |  Low         |  High                         |
| Example      | `int a[10]`  | `ArrayList`, `vector`, `list` |


5. Time Complexity Comparison

| Operation       | Static Array  | Dynamic Array  |
| --------------- | ------------- | -------------- |
| Access          | O(1)          | O(1)           |
| Update          | O(1)          | O(1)           |
| Insert (end)    | Not allowed   | O(1) amortized |
| Insert (middle) | O(n)          | O(n)           |
| Delete          | O(n)          | O(n)           |
